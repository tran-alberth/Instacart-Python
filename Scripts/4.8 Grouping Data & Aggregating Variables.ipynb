## Use groupby()
#### 1. Split the data into groups based on some criteria
#### 2. Apply a function to each group separately
#### 3. Combine the results into a dataframe or alternative data structure or create a new column in the current dataframe

```python
# 1. Spliting data into groups based on "department_id"
# 2. Applying agg() function to each group to obtain the mean values for the "order_number" column

df.groupby('department_id').agg({'order_number' : ['mean']})
```

```python
# Alternative:

df.groupby('department_id')['order_number'].mean()
```

#### When using agg(), put the column targeted inside parentheses of agg() function as argument.
#### When using mean() (or any other standard aggregation function), simply index column with brackets & follow with function used.

```python
# Another alternative:

df.groupby('department_id').order_number.mean()
```

## Performing multiple aggregations

```python
df.groupby('department_id').agg({'order_number' : ['mean','min','max']})
```

## Aggregating Data with transform()

```python
# By using 'user_id', we transform 'order_number' to generate maximum orders for each user into new 'max_order' column

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

## Creating loyalty flag using loc

```python
ords_prods_merge.loc[ords_prods_merge['max_order']>40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order']<=40) & (ords_prods_merge['max_order']>10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order']<=10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

```python
# Transforming data in order to create spending flag

ords_prods_merge['mean_prices'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.max)
```

```python
# Creating spending_flag

ords_prods_merge.loc[ords_prods_merge['mean_prices']<10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['mean_prices']>=10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge[['user_id', 'mean_prices', 'spending_flag']].head(60)
```

```python
# Creating frequency flag

ords_prods_merge['median_days_since_previous_order'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_previous_order']>20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_since_previous_order']>10) & (ords_prods_merge['median_days_since_previous_order']<=20), 'frequency_flag'] = 'Regular spender'
ords_prods_merge.loc[ords_prods_merge['median_days_since_previous_order']<=10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```
